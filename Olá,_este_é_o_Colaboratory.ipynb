```
import numpy as np
import matplotlib.pyplot as plt

# Exemplo de dados
X = np.array([2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022])
y = np.array([419.30,416.84,424.92,428.53,457.84,428.30,434.0,444.84,553.15,731.15,521.38])

# Adicionando uma coluna de 1s para o termo de intercepto
X_b = np.c_[np.ones(X.shape[0]), X]

# Calculando os coeficientes usando a fórmula dos mínimos quadrados
# beta = (X_b.T * X_b)^(-1) * X_b.T * y
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

# Coeficientes
intercept = beta[0]
slope = beta[1]

print(f"Intercepto: {intercept}")
print(f"Coeficiente (slope): {slope}")

# Fazendo previsões
y_pred = X_b @ beta

# Plotando os dados e a linha de regressão
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X, y_pred, color='red', label='Linha de regressão')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

# Seus dados
X = np.array([2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022])
y = np.array([419.30, 416.84, 424.92, 428.53, 457.84, 428.30, 434.0, 444.84, 553.15, 731.15, 521.38])

# Adicionando uma coluna de 1s para o termo de intercepto
X_b = np.c_[np.ones(X.shape[0]), X]

# Calculando os coeficientes usando a fórmula dos mínimos quadrados
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

# Coeficientes
intercept = beta[0]
slope = beta[1]

print(f"Intercepto: {intercept:.2f}")
print(f"Coeficiente (slope): {slope:.2f}")

# Fazendo previsões para os dados existentes
y_pred = X_b @ beta

# Cálculo de R^2
ss_total = np.sum((y - np.mean(y)) ** 2)  # Soma total dos quadrados
ss_residual = np.sum((y - y_pred) ** 2)  # Soma dos quadrados dos resíduos
r_squared = 1 - (ss_residual / ss_total)

# Cálculo do RMSE
rmse = np.sqrt(np.mean((y - y_pred) ** 2))

print(f"R²: {r_squared:.4f}")
print(f"RMSE: {rmse:.2f}")

# Prevendo até 2024
X_future = np.array([2023, 2024])
X_future_b = np.c_[np.ones(X_future.shape[0]), X_future]  # Adicionando coluna de 1s
y_future = X_future_b @ beta  # Previsões

# Fazendo previsões para todos os anos
X_all = np.concatenate((X, X_future))
y_pred_all = np.concatenate((y_pred, y_future))

# Plotando os dados e a linha de regressão
plt.scatter(X, y, color='blue', label='Dados reais')
plt.plot(X_all, y_pred_all, color='red', label='Linha de regressão')
plt.scatter(X_future, y_future, color='green', marker='x', label='Previsões')
plt.xlabel('Ano')
plt.ylabel('Valor')
plt.title('Regressão Linear com Previsões até 2024')
plt.legend()
plt.grid()
plt.show()
```
